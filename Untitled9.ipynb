```python
import numpy as np
import matplotlib.pyplot as plt

# Reproducibility
np.random.seed(42)

# Generate data
m = 200
X = np.random.uniform(0, 5, m)
eps = np.random.normal(0, 1, m)
y = 3 + 4*X + eps

# Design matrix with bias
X_b = np.c_[np.ones((m,1)), X]  # shape (m,2)

# Closed-form solution
theta_closed = np.linalg.inv(X_b.T @ X_b) @ (X_b.T @ y)
print("Closed-form intercept:", theta_closed[0])
print("Closed-form slope:", theta_closed[1])

# Gradient Descent
theta = np.zeros(2)
eta = 0.05
iters = 1000
losses = []

for i in range(iters):
    yhat = X_b @ theta
    r = yhat - y
    grad = (2/m) * (X_b.T @ r)
    theta -= eta * grad
    losses.append(np.mean(r**2))

print("GD intercept:", theta[0])
print("GD slope:", theta[1])

# Plot data + fitted lines
plt.scatter(X, y, alpha=0.6, label="Data")
# Use sorted X for cleaner lines
order = np.argsort(X)
X_sorted = X[order]
Xb_sorted = X_b[order]
y_closed = Xb_sorted @ theta_closed
y_gd = Xb_sorted @ theta

plt.plot(X_sorted, y_closed, label="Closed-form")
plt.plot(X_sorted, y_gd, linestyle='--', label="Gradient Descent")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Linear Regression: Closed-form vs Gradient Descent")
plt.show()

# Loss curve
plt.plot(range(iters), losses)
plt.xlabel("Iteration")
plt.ylabel("MSE")
plt.title("Loss vs Iterations (GD)")
plt.show()

# Short explanation
print("\nExplanation: Both solutions converge to nearly the same parameters (~intercept=3, slope=4). "
      "GD takes iterative steps guided by the gradient and approaches the closed-form solution.")
```
